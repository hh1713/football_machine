##  Creating the teams dataframe

```python
##Define the object for the season team
class MyDf:
    def __init__(self, mydf):
        self.df = mydf
```

```python
# reading relevant data (skipping referee and betting data)
df = pd.read_csv("../Data/EPL_1617.csv", usecols=range(1,10))
```

```python
df.head()
```

```python
teams = df.HomeTeam.unique()
```

```python
allteamsresults = []
for i in teams:
#     path=pd.DataFrame()            
    path = df[(df["HomeTeam"] == i) | (df['AwayTeam'] == i )] 
    allteamsresults.append(MyDf(path))
```

## allteamsresults is a list of dataframes and each dataframe represents a specific team. To access info on a team use team index. The position of the team in the list "teams" will give you the index of the team in the "allteamsresults" list of dataframes.

## Calculating data for teams

```python
for index, name in enumerate(teams):
    print index
```

```python
def calculate_for_team_i(team_df, team_name, team_index):
    initial_value = 0
    for i in team_df[team_index].df.index:
        if (team_df[team_index].df.HomeTeam[i] == team_name):# and team_df[team_index].df.FTR[i] == 'H'):
            team_df[team_index].df['total_points'] += 3
        elif (team_df[team_index].df.AwayTeam[i] == team_name):# and team_df[team_index].df.FTR[i] == 'A'):
            team_df[team_index].df['total_points'] += 3
        else:
            team_df[team_index].df['total_points'] += 1
    return team_df[team_index].df
```

```python
for index, name in enumerate(teams):
    for i in allteamsresults[index].df.index:
        if (allteamsresults[index].df.HomeTeam[i] == name):
            allteamsresults[index].df.home_team_winner[i] = 'yes'
        else:
            allteamsresults[index].df.home_team_winner[i] = 'no'
```

```python
for index, name in enumerate(teams):
    allteamsresults[index].df = calculate_for_team_i(allteamsresults, name, index)
```
